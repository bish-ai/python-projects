```python
class Employee:
  def __init__(self,name,salary,emp_id):
    self.name=name
    self.salary=salary
    self.emp_id=emp_id
  def display(self):
    print("name:",self.name)
    print("salary:",self.salary)
    print("emp_id",self.emp_id)
```

```python
employees=[]
while True:
  print("\n ===MENU===")
  print("1.add employees")
  print("2.display employees")
  print("3.Exit")
  choice=int(input("enter your choice:"))
  if choice ==1:
    emp_id=int(input("enter id:"))
    name=input("enter the name:")
    salary=input("enter the salary:")
    emp=Employee(emp_id,name,salary)
    employees.append(emp)
    break
  elif choice ==2:
    if len(employees)==0:
        print("no employees are found")
    else:
      for emp in employees:
        emp.display()
        break
  elif choice ==3:
    print("existing.....")
    break
  else:
    print("invalid choice")
```
